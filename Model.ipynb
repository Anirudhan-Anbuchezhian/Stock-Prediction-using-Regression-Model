# Linear Regression

---

```python
                        
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from math import sqrt
from sklearn.model_selection import train_test_split


# getting all the apple stock data from the csv file and storing it into dataframe named df
df = pd.read_csv('AAPL.csv')


#takes open price as x value and close price as y value
#i = len(df['Date'])
#x = np.array(range(1,(i+1)))
x = df['Open'].values
y = df['Close'].values

# Partition data into training and testing set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,shuffle=False)

#reshaping the data into 2-D array to apply linear regression
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)


# creating a Linear Regression model
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

#ploting the predicted values with the actual values
plt.title('Linear Regression')
plt.xlabel('Opening Price of AAPL Stock')
plt.ylabel('Closing Price of AAPL Stock')
plt.scatter(x_test,y_test)
plt.plot(x_test, y_pred)
plt.show()

#finding the accurecy of the prediction
print("R2: ",metrics.r2_score(y_test, y_pred))
print("RMSC: ",sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("MSE: ",metrics.mean_squared_error(y_test, y_pred))
print("MAE: ",metrics.mean_absolute_error(y_test, y_pred))

```

# Polynomial Regression

---

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import sqrt
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures

# getting all the apple stock data from the csv file and storing it into dataframe named df
df = pd.read_csv('AAPL.csv')


#takes open price as x value and close price as y value
#i = len(df['Date'])
#x = np.array(range(1,(i+1)))
x = df['Open'].values
y = df['Close'].values

#Partition data into training and testing set
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,shuffle=False)

#reshaping the data into 2-D array to apply linear regression
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)

# Setting features for polynomial regression
polynomial_features= PolynomialFeatures(degree=3)
x_test_poly = polynomial_features.fit_transform(x_test)
x_train_poly = polynomial_features.fit_transform(x_train)

# creating a polynomial regression using Linear regression model
model = LinearRegression()
model.fit(x_train_poly, y_train)
y_pred = model.predict(x_test_poly)

#ploting the predicted values with the actual values
plt.title('Polynomial Regression')
plt.xlabel('Opening Price of AAPL Stock')
plt.ylabel('Closing Price of AAPL Stock')
plt.scatter(x_test, y_test)
plt.plot(x_test, y_pred)
plt.show()

#finding the accurecy of the prediction
print("R2: ",metrics.r2_score(y_test, y_pred))
print("RMSC: ",sqrt(metrics.mean_squared_error(y_test, y_pred)))
print("MSE: ",metrics.mean_squared_error(y_test, y_pred))
print("MAE: ",metrics.mean_absolute_error(y_test, y_pred))
```
